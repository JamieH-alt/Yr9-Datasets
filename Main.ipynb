### Privacy And Security
* Application Data Privacy: \
 = My application does not collect any data from the user. And everything searched is stored on the users computer. The closest to identifiable data that the program requires is a city that the user inputs to pull the data from. There are also zero foreseeable changes/responsibilites that I'd need to make if I was to push this application to the public.
* Cyber Security: \
 = A normal application should have many things in order to maintain the security of its users. It should have User Authentication, Password Hashing and Encryption to keep all personally identifiable details of the user safe. \
 = User Authentication is the process in which a persons identified is verified before they are allowed access to an online service.
 = Password Hashing is where passwords are changed mathematically before being stored. There's no way to reverse the process and the code for it is stored on the users application. When User Authentication occurs it checks a hashed password against the hashed password saved in the server; allowing a process in which the password itself is never sent across the internet. \
 = Encryption is a way to scramble/change/mix data so that only authorised parties with an "Encryption Key" can unscramble and read it.
