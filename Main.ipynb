### Key Details

* Data: \
 = I am using Open Meteo's data API to look at the temperature over the course of a year. (If possible) 
* Goal: \
 = To identify the coldest days and warmest days to see when seasons mainly pick up (As they have a bit of overlap, e.g. Summer starts cold, Winter starts warm) 
* Source: \
 = https://open-meteo.com/en/
* Access: \
 = Public as long as you know how to use an API.
* Access Method: \
 = Via its online API service.

---

### Functional Requirements
* Loading \
 = The API needs to let me obtain large amounts of data as I'm attempting to look at the weather of a year period. And it all needs to be formatted correctly so it loads easily into pandas.
* Data Cleaning \
 = The system shouldn't need any data cleaning as it appears to be well maintained
* Data Analysis \
 = The system needs to allow for mean averaging (And potentially linear regression if I have the time.)
* Data Visualisation \
 = The data visualisation has not yet been finalised but it will most likely be through Matplotlib or some other similar aplication.
* Data Reporting \
 = The system should output the charts to the screen and it shouldn't need to store anything as a TXT unless needed.

---

 ### Use Cases
* Loading \
--  Actor: User \
--  Goal: To choose which dataset to load into the system \
--  Preconditions: User knows the format in which this needs to be inputed (Name of place) \
--  Main Flow: \
------  1. User inputs the location into the right spot in the User interface. \
------  2. System validates the link. \
------  3. System loads the dataset and proccesses into a chart. \
--  Postconditions: Dataset is loaded and shown to the user as a chart.
* Data Reporting \
--  Actor: User \
--  Goal: To decide whether or not to save a file, and to name that file. \
--  Preconditions: User has already loaded an data using an API call from the `Loading` Section \
--  Main Flow: \
------  1. User inputs a file name into the selection box \
------  2. User hits save. \
--  Postconditions: Dataset is saved as a file with the name input by the User.
